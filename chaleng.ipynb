```python
medala_list_we = list()
medala_list_hei = list()
medala_list_age = list()
No_medala_list_we = list()
No_medala_list_hei = list()
No_medala_list_age = list()
for i in range(len(age_list)):
    if df.loc[i]["Medal"] != "No Medal":
        medala_list_we.append(df.loc[i]["Weight"])
        medala_list_hei.append(df.loc[i]["Height"])
        medala_list_age.append(df.loc[i]["Age"])
    else:
        No_medala_list_we.append(df.loc[i]["Weight"])
        No_medala_list_hei.append(df.loc[i]["Height"])
        No_medala_list_age.append(df.loc[i]["Age"])
```

```python
print(len(medala_list_we) == len(medala_list_hei))
print(len(No_medala_list_hei) == len(No_medala_list_age))
```

```python
x = [(medala_list_we[i] + medala_list_hei[i]) * 2**-2 * (medala_list_age[i]) for i in range(len(medala_list_we))]
y = [(medala_list_we[i] * medala_list_hei[i]) * 2**-2 * (medala_list_age[i]) for i in range(len(medala_list_we))]
z = [(2**-2) * x[i] * y[i] for i in range(len(medala_list_we))]

nx = [(No_medala_list_we[i] + No_medala_list_hei[i]) * 2**-2 * (No_medala_list_age[i]) for i in range(len(No_medala_list_we))]
ny = [(No_medala_list_we[i] * No_medala_list_hei[i]) * 2**-2 * (No_medala_list_age[i]) for i in range(len(No_medala_list_we))]
nz = [(2**-2) * nx[i] * ny[i] for i in range(len(No_medala_list_we))]
```

```python
ax = plt.axes(projection = '3d')
ax.scatter3D(x,y,z , color = "green")
ax.scatter3D(nx,ny,nz , color = "red")
plt.show()
```

```python
sport = list(set(list(df["Sport"])))
sport
```

```python
sport_di = dict()
sport_di_number = dict()
for i in sport: sport_di[i] = [0,0,0]
for i in sport: sport_di_number[i] = 0
```

```python
sport_di
```

```python
df = df.reset_index(drop = True)
for i in range(len(df)):
    sport_di_number[df.loc[i]["Sport"]] += 1
    sport_di[df.loc[i]["Sport"]][0] += df.loc[i]["Age"]
    sport_di[df.loc[i]["Sport"]][0] += df.loc[i]["Height"]
    sport_di[df.loc[i]["Sport"]][0] += df.loc[i]["Weight"]
```

```python
sport_di_number
```

```python
for i in sport_di:
    sport_di[i] = sum(sport_di[i]) / sport_di_number[i]
```

```python
sport_di
```

```python
def unaliz():
    name = input("What is your name? ")
    while True:
        temp = True
        age = input("what is your age?: ")
        for i in age:
            if i not in ["1","2","3","4","5","6","7","8","9",".","0"]: temp = False
        if temp == True:
            if float(age) > 14 and float(age) < 110:
                age = float(age)
                break
    while True:
        temp = True
        height = input("what is your height?: ")
        for i in height:
            if i not in ["1","2","3","4","5","6","7","8","9",".","0"]: temp = False
        if temp == True:
            if float(height) > 50 and float(height) < 250:
                height = float(height)
                break
    while True:
        temp = True
        weight = input("what is your weght?: ")
        for i in weight:
            if i not in ["1","2","3","4","5","6","7","8","9",".","0"]: temp = False
        if temp == True:
            if float(weight) > 20 and float(weight) < 120:
                weight = float(weight)
                break

    mid = sum([weight,height,age]) / 3
    list_mid = list()
    for i in sport:
        temp_number = sport_di[i] - mid
        if temp_number < 0: temp_number *= -1
        list_mid.append(temp_number)
    temp_darsad = (100 * mid) / float(sport_di[sport[list_mid.index(min(list_mid))]])
    if temp_darsad > 80: temp_medal = "Gold"
    elif temp_darsad > 50: temp_medal = "Silver"
    elif temp_darsad > 30: temp_medal = "Brownz"
    else: temp_medal = "No Medal"
    if temp_medal != "No Medal": print(f"Hello {name}\nYou will most likely win the {sport[list_mid.index(min(list_mid))]} {temp_medal} medal and your success rate is {temp_darsad}%.")
    else: print(f"Your suitable sport is {sport[list_mid.index(min(list_mid))]}, but the probability of winning a medal is less than 30%  ({temp_darsad}%).")
    
```
